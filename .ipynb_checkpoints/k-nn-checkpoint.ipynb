## k-NN(k-최근접 이웃)

- 가장 가까운 훈련 데이터 초인트 하나를 최근접 이웃으로 찾아 예측에 사용한다. 
- 훈련 데이터셋을 그냥 저장하는 것이 모델을 만드는 것 = 가장 간단한 머신러닝 알고리즘 

```python
import mglearn
import matplotlib.pyplot as plt
import warnings

#warning 메세지 무시 
warnings.filterwarnings('ignore')

# 데이터 셋 만들기 
X, y = mglearn.datasets.make_forge()

mglearn.discrete_scatter(X[:,0], X[:,1],y)
plt.legend(["class 0","class 1"], loc=4)
plt.xlabel("first feature")
plt.xlabel("second feature")
print("X.shape : ",X.shape)
```

```python
mglearn.plots.plot_knn_classification(n_neighbors=1)  #forge data에 대한 1-최근접 이웃 모델 예측 
#데이터 포인트 3개에 대해 가장 가까운 훈련 데이터 포인트를 연결 -> 결과 : 연결된 훈련 데이터 포인트와 같은 label로 예측 
```

```python
#k개의 이웃에 대한 알고리즘  : k-nn 알고리즘 
#레이블을 정하기 위해 투표를 하여 이웃이 더 많은 클래스의 label로 예측, 즉 k개의 이웃중 다수의 클래스가 label이 된다. 
mglearn.plots.plot_knn_classification(n_neighbors=3) # k=3
```

### 클래스가 여러개인 데이터셋에도 같은 방법 적용 
## scikit-learn을 이용한 k-nn 알고리즘

### 1. 데이터 나누기

```python
from sklearn.model_selection import train_test_split
X,y = mglearn.datasets.make_forge()

X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)
```

### 2. KNeighborsClassifier를 import 하고 객체만들기 

```python
#이웃 수 지정 (매개변수)
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)
```

```python
#훈련 셋을 사용하여 모델 학습시키기
clf.fit(X_train, y_train)
```

```python
#predic메소드 호출 
#테스트 셋의 각 데이터 포인트에 대해 훈련 셋에거 가장 가까운 이웃을 계산한 다음 가장 많은 클래스 찾기 
print("테스트 데이터 예측: ",clf.predict(X_test))
```

```python
# 모델 평가 
print("테스트 셋의 정확도 측정 : {:.2f}".format(clf.score(X_test, y_test)))
```

## KNeighborsClassifier 분석
- 알고리즘이 클래스 0과 클래스 1로 지정한 영역으로 나뉘는 결정경계를 볼 수 있다. 

```python
#k가 1,3,9 일때의 결정경계를 보여준다. 
fig, axes = plt.subplots(1,3,figsize = (10,3))

for n_neighbors, ax in zip([1,3,9], axes):
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X,y)
    mglearn.plots.plot_2d_separator(clf,X, fill = True, eps = 0.5, ax= ax, alpha=.4)
    mglearn.discrete_scatter(X[:,0], X[:,1],y,ax=ax)
    ax.set_title("{} neighbor".format(n_neighbors))
    ax.set_xlabel("feature 0")
    ax.set_ylabel("feature 1")

axes[0].legend(loc=3)
```

- 이웃의 수가 늘어날 수록 결정경계 (decision boundary)가 더 부드러워지는 것을 알 수 있다.
- 부드러운 경계는 더 단순한 모델을 의미한다. 
- 이웃을 적게 잡을수록 모델의 복잡도가 높아진다. 


## 모델의 복잡도와 일반화 사이의 관계 

```python
#실제 유방암 데이터셋 ㅇ용

from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, stratify=cancer.target, random_state=66)

training_accuracy = []
test_accuracy = []

neighbors_settings = range(1,11)
```

```python
# 모델 생성
for n_neighbors in neighbors_settings:
    clf = KNeighborsClassifier(n_neighbors= n_neighbors)
    clf.fit(X_train, y_train)
    training_accuracy.append(clf.score(X_train, y_train))#훈련 셋 정확도 저장
    test_accuracy.append(clf.score(X_test, y_test)) #일반화 정확도 저장 
    
   
plt.plot(neighbors_settings, training_accuracy, label="train_acc")
plt.plot(neighbors_settings, test_accuracy, label="test_acc")
    
plt.ylabel("accuracy")
plt.xlabel("n_neighbors")
plt.legend()
    
    
    
```

- 이웃수에 따른 훈련 셋과 테스트셋의 정확도를 보여준다. 
- 테스트 셋의 정확도가 가장 좋을 때는 여섯개를 사용한 경우 
- 이웃의 수가 늘어나며 모델이 단순해지고 훈련 데이터의 정확도는 줄어든다. 


## K-최근점 이웃 회귀
- k-nn 알고리즘의 경우 회귀에도 사용된다. 


```python
#Wave 데이터셋 만들기 - 입력 feature 하나, 모델링할 target 
X, y = mglearn.datasets.make_wave(n_samples=40)
plt.plot(X,y,'o')
plt.ylim(-3,3)
plt.xlabel('feature')
plt.ylabel('target')
```

```python
mglearn.plots.plot_knn_regression(n_neighbors=1)   #k=1
```

```python
mglearn.plots.plot_knn_regression(n_neighbors=2)   #k=2
```

```python
from sklearn.neighbors import KNeighborsRegressor

X,y = mglearn.datasets.make_wave(n_samples=40)
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,y,random_state=0)

#k=3
reg = KNeighborsRegressor(n_neighbors=3)

reg.fit(X_train,y_train)
```

```python
print("Test result : \n", reg.predict(X_test))
```

```python
#score 메소드를 이용해 모델 평가하기 
#회귀의 경우 R^2 값을 반환, 외귀 모델에서 예측의 적합도를 측정한것 (0~1 사이의 값) 1:예측이 완벽한 경우, 0:trainset의 출력값인 y_train의 평균으로만 예측
#R^2은 음수가 될수도 있다 -> 예측과 target이 반대의 결과를 가지는 경우 

print("test set R^2 : {:.2f}".format(reg.score(X_test, y_test)))


```

## KNeighborsRegressor 분석

1차원 데이터셋에 대해 가능한 모든 특성값을 만들어서 예측가능

```python
import numpy as np

fig, axes = plt.subplots(1,3,figsize = (15,4))

line = np.linspace(-3,3,1000).reshape(-1,1)

for n_neighbors, ax in zip([1,3,9], axes):
    reg = KNeighborsRegressor(n_neighbors=n_neighbors)
    reg.fit(X_train,y_train)
    ax.plot(line, reg.predict(line))
    ax.plot(X_train, y_train, '^', c=mglearn.cm2(0),markersize=8)
    ax.plot(X_test, y_test, 'v', c=mglearn.cm2(1),markersize=8)    
    
    ax.set_title("{} neighbors train score : {:.2f}   test score : {:.2f}".format(n_neighbors, reg.score(X_train,y_train),
                                                                                  reg.score(X_test,y_test)))
    ax.set_xlabel("feature")
    ax.set_ylabel("target")
axes[0].legend(["model predict", "train data/target", "test data/target"],loc ="best" )
```

- 이웃이 하나일때  : 훈련 세트의 각 sample이 predict에 주는 영향이 큼 = > 예측값이 모두 train data를 지나감 

- 이웃이 많을 때 : 훈련 데이터에는 잘맞지 않더라도 더 안정된 예측값을 얻을 수 있음 


## Knn의 특징

### KNeighbors 분류기에서 중요한 매개변수 
- sample 사이의 거리를 재는 방법 (일반적으로 유클리디안 거리 방식 적용)
- 이웃의 수 (실제로 이웃의 수는 3,5개 정도로 작을 때 잘작동한다.)

### 특징
- 이해하기 쉬운 모델 
- 많은 조정없이도 좋은 성능을 보임 
- 복잡한 알고리즘 사용 전 시도해볼만함 
- 훈련 셋의 크기가 커지면 예측이 느려진다
- K-nn 사용시 모델의 전처리 과정이 매우 중요해진다
- 많은 특성을 가진 데이터셋에서는 잘 작동하지 않는다 
- 특성값의 대부분이 0인 (희소한) 데이터셋에서는 특히 성능이 낮음 

- 예측이 느리고 많은 특성을 처리하기에는 어려움이 존재함 -> 현업에서는 대부분 사용하지 않음
